# Análise Explorátória 

```python
#1- Carregar as Bibliotecas Numpy, pandas, plotly, matplotlib...

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly
import plotly.plotly as py
import plotly.offline as py
import plotly.graph_objs as go
import seaborn as sns
```

```python
#2- Verificar e Interpretar as variáveis Head, dtypes, describe.

crime = pd.read_csv('../analise_exploratoria/crimeZip/denver-crime-data/crime.csv');
crime.head().T
```

```python
# 3 quero apenas os crimes
df_crime = crime[crime['IS_CRIME'] == 1]
```

```python
# 4 Analisar as informações dos dados para ver os tipos, tamanho e médias se for o caso
df_crime.info()
#df_crime.shape
#df_crime.describe()
```

#  1- Quero saber quais são os horários de maior frequência de crimes? 

```python
# tranformar o campo FIRST_OCCURRENCE_DATE para data

df_crime['FIRST_OCCURRENCE_DATE'] = pd.to_datetime(df_crime['FIRST_OCCURRENCE_DATE'],  infer_datetime_format=True)
```

```python
# Pegar apenas a hora
df_crime['Hora'] = df_crime['FIRST_OCCURRENCE_DATE'].dt.strftime('%H')
```

```python
# transformar a hora em númerico pra plotar os dados
df_crime['Hora'] = pd.to_numeric(df_crime['Hora'])
```

```python
## 1ª análise, nela consta a relação dos horários e sua frequência de crime é um histograma
ax = df_crime.Hora.plot.hist(bins=24);
ax.set_xlabel('Hora')
ax.set_ylabel('Frequencia')
```

#  2- Como está distribuído a quantidade de crime por hora para cada distrito?

```python
## 2ª análise, a quantidade de crime por hora e distrito

lista = ['Crime Regiao 1','Crime Regiao 2','Crime Regiao 3','Crime Regiao 4','Crime Regiao 5','Crime Regiao 6','Crime Regiao 7']

axes = df_crime.pivot_table(index='Hora', columns='DISTRICT_ID', values='OFFENSE_TYPE_ID', aggfunc='count').plot(kind='bar', subplots=True,  title=lista,  figsize=(20,40), sharex=False)
plt.subplots_adjust(wspace=0.5, hspace=0.5)

axes[0].set_ylabel('frequencia')
axes[1].set_ylabel('frequencia')
axes[2].set_ylabel('frequencia')
axes[3].set_ylabel('frequencia')
axes[4].set_ylabel('frequencia')
axes[5].set_ylabel('frequencia')
axes[6].set_ylabel('frequencia')
```

# 3- Quais são as categorias de crimes que mais ocorrem?

```python
## 3ª análise, Quantidade de crimes por catégoria, Afinal seria importante saber quais são os tipos de crimes que mais ocorrem né.

plt.subplots(figsize=(10,4))

order = df_crime['OFFENSE_CATEGORY_ID'].value_counts().index
sns.countplot(df_crime['OFFENSE_CATEGORY_ID'],order=order, color="blue")

plt.xticks(rotation = 90)
plt.title("Quantidade de crimes por categoria")
plt.xlabel(""); 
plt.ylabel("Frequência"); 
```

# 4- Será que a quantidade de crime tende a aumentar durante os anos?


```python
# Pegar apenas o ano que ocorreu o crime
df_crime['ANO'] = df_crime['FIRST_OCCURRENCE_DATE'].dt.strftime('%Y')

```

```python
plt.subplots(figsize=(10,4))

sns.countplot(df_crime['ANO'], color="blue")

plt.xticks(rotation = 90)
plt.title("Quantidade de crimes por Ano")
plt.xlabel(""); 
plt.ylabel("Frequência"); 
```

# Como 2019 ainda não terminou ( o Dataset só tem informações completas até 2018), portanto vamos ignorar o ano de 2019 para esta análise!

```python
df_crime_anos = df_crime[df_crime['ANO'] < '2019'] 
```

```python
plt.subplots(figsize=(10,4))

sns.countplot(df_crime_anos['ANO'], color="blue")

plt.xticks(rotation = 90)
plt.title("Quantidade de crimes por Ano")
plt.xlabel(""); 
plt.ylabel("Frequência"); 
```

# 5- Quantidade de Crimes ao decorrer dos meses por ano, será que algum mês tem alguma tendência de ter mais ou menos crimes?

```python
df_crime['MES'] = pd.DatetimeIndex(df_crime['FIRST_OCCURRENCE_DATE']).month
```

```python
plt.subplots(figsize=(14,5))
plt.plot(df_crime[['MES','OFFENSE_CATEGORY_ID']][df_crime['ANO']=='2014'].groupby('MES').agg('count'))
plt.plot(df_crime[['MES','OFFENSE_CATEGORY_ID']][df_crime['ANO']=='2015'].groupby('MES').agg('count'))
plt.plot(df_crime[['MES','OFFENSE_CATEGORY_ID']][df_crime['ANO']=='2016'].groupby('MES').agg('count'))
plt.plot(df_crime[['MES','OFFENSE_CATEGORY_ID']][df_crime['ANO']=='2017'].groupby('MES').agg('count'))
plt.plot(df_crime[['MES','OFFENSE_CATEGORY_ID']][df_crime['ANO']=='2018'].groupby('MES').agg('count'))
plt.legend(['2014','2015','2016','2017','2018'])

plt.xlabel('Mês')
plt.ylabel('Frequência');
```
