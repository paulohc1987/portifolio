```python
X = train_labels.drop(['city','year','weekofyear','week_start_date','total_cases','month'], axis=1)
y = train_labels['total_cases']
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
import auto_lgbm
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
xb = XGBRegressor()
lb = LGBMRegressor()
```

```python
xb.fit(X_train, y_train)
y_pred_xb = xb.predict(X_test)
```

```python
xb.fit(X_train, y_train)
y_pred_xb = xb.predict(X_test)

lb.fit(X_train, y_train)
y_pred_lb = lb.predict(X_test)
```

```python
y_pred_xb
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred_xb)), np.sqrt(mean_squared_error(y_test,y_pred_lb)))
```
